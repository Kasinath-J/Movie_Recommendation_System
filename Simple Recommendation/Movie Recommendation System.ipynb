```python
meanvote = movies['vote_average'].mean()
print(meanvote)
```

```python
minvote = movies['vote_count'].quantile(0.80)
print(minvote)
```

```python
q_movies = movies.copy().loc[movies['vote_count'] >= minvote]
```

```python
q_movies.shape
```

```python
def weighted_rating(x,minvote = minvote,meanvote=meanvote):
    voters = x['vote_count']
    avg_vote = x['vote_average']
    return(voters/(voters+minvote)*avg_vote)+(minvote/(minvote+voters)*meanvote)
```

```python
q_movies['score'] = q_movies.apply(weighted_rating,axis=1)
```

```python
q_movies = q_movies.sort_values('score',ascending = False)
```

```python
def findGenre(x):
    lsts = []
    genre = x['genres']
    for i in genre:
        lst = []
        temp = eval(i)
        for j in temp:
            lst.append(j['name'])
        lsts.append(lst)
    return lsts
```

```python
genreset = set()
for i in genreLst:
    for j in i:
        genreset.add(j)

availableGenres = list(genreset)
availableGenres.sort()
```

```python
genreDict = {}
for i in range(1,len(availableGenres)):
    genreDict[i] = availableGenres[i]
genreDict
```

```python
def pickTopMovies(x,inp):
    genres = x['genreLst']
    lst1= []
    for i in range(len(genres)):
        for j in genres[i]:
            if(inp == j):
                lst1.append(x.iloc[i]['title'])
    df = pd.DataFrame(lst1,columns = ['Recommendations'])
    df.index += 1
    print(df.head(15))
```
