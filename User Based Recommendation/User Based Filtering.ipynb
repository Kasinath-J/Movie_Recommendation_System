```python
import pandas as pd
import numpy as np
```

```python
movies_ratings = pd.read_csv(r"C:\Users\abhin\Downloads\ratings.csv",nrows = 1048576)
print(movies_ratings)
```

```python
dup_data = movies_ratings.copy()
print(dup_data)
```

```python
final_dataset = movies_ratings.pivot_table(index = 'userId', columns = 'movieId', values = 'rating').fillna(0)
```

```python
print(final_dataset)
```

```python
dup_data['rating'] = dup_data['rating'].apply(lambda x: 0 if x > 0 else 1)
```

```python
dup_final = dup_data.pivot_table(index = 'userId', columns = 'movieId', values = 'rating').fillna(1)
print(dup_final)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(final_dataset)
similarity[np.isnan(similarity)] = 0
print(similarity)
```

```python
user_predicted_ratings = np.dot(similarity, final_dataset)
print(user_predicted_ratings)
```

```python
user_final_ratings = np.multiply(user_predicted_ratings,dup_final)
print(user_final_ratings.head())
```
